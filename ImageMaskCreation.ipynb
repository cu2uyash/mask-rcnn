```python
import cv2
import numpy as np
from skimage.draw import circle
import glob
import os
import scipy as sp
```

```python
folder = "C:\MachineLearning\FemurHead\Data-backup - Copy"

rootdir = folder
return_list = []
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        #print (os.path.join(subdir, file))
        filepath = subdir + os.sep + file
        if filepath.endswith('.yaml'):
            return_list.append(filepath)
            
len(return_list)
            
```

```python
return_img_lst = []

for path in return_list:
    path = os.path.splitext(path)[0]+"_mf.tif"
    return_img_lst.append(path)

len(return_img_lst)
```

```python
write_img_lst = []

for path in return_list:
    path = os.path.splitext(path)[0]+"_label.png"
    write_img_lst.append(path)

len(write_img_lst)
```

```python

def SaveMask(rpath,wpath,xcoord,ycoord,radius) :
    img = cv2.imread(rpath,0)
    height= img.shape[0]
    width = img.shape[1]
    print(rpath)
    print(height)
    print(width)
    print(ycoord)
    print(xcoord)
    print(radius)
    blank_image = np.zeros((height,width),dtype=np.uint8)
    rr, cc = circle(ycoord,xcoord, radius, shape=(height, width))
    blank_image[rr, cc] = 1

    cv2.imwrite(wpath,blank_image)
    
```

```python
count=0
import yaml
for slist in return_list:
    with open(slist) as f:
        contents = yaml.load(f)
        x_cord = contents['femoral_head']['manual_head_x_orig']
        y_cord = contents['femoral_head']['manual_head_y_orig']
        radius = contents['femoral_head']['manual_head_r_orig']
        
    SaveMask(return_img_lst[count],write_img_lst[count],float(x_cord),float(y_cord),float(radius))
    count+=1
```
